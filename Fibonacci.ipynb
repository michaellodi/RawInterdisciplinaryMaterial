```python
def fibo1(n):
    if n<2:
        return n
    global count_ric
    count_ric += 1
    return fibo1(n-1)+fibo1(n-2)

def fibo2(n):
    a,b= (1,0)
    while n>1:
        global count_iter
        count_iter += 1
        a, b = a+b, a
        n -= 1
    return a


#Calculations below to count steps, time and realize graphs. 
#Not necessary to be understood by students.

from time import time


L_count_ric = []
L_count_iter = []
L_time_ric = []
L_time_iter = []

for i in range(40):
    
    count_ric = 0
    count_iter = 0
    
    start_ric = time()
    fibo1(i)
    total_ric = time()-start_ric
    
    start_iter = time()
    fibo2(i)
    total_iter = time()-start_iter
    
    L_count_ric.append(count_ric)
    L_count_iter.append(count_iter)
    L_time_ric.append(total_ric)
    L_time_iter.append(total_iter)
    
    #print(i,count_ric, count_iter)
    #print(i, f'{total_ric:.10f}', f'{total_iter:.10f}')
    
    
    
from matplotlib import pyplot as plt
plt.plot(range(40), L_count_iter, label="fibo2 (iterative)")
plt.plot(range(40),L_count_ric, label="fibo1 (recursive)")
plt.title("Number of sums")
plt.legend()
plt.show()

plt.plot(range(40), L_time_iter, label="fibo2 (iterative)")
plt.plot(range(40),L_time_ric, label="fibo1 (recursive)")
plt.title("Seconds elapsed")
plt.legend()
plt.show()

plt.plot(range(40), L_count_iter, label="fibo2 (iterative)")
plt.title("Number of sums")
plt.legend()
plt.show()

plt.plot(range(40), L_time_iter, label="fibo2 (iterative)")
plt.title("Seconds elapsed")
plt.legend()
plt.show()
```
